- Графіки та візуалізація: Вбудована підтримка для графіків та візуалізацій за допомогою бібліотек, таких як Matplotlib, Seaborn, Plotly тощо.  


```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, label='sin(x)')
plt.xlabel('x'), plt.ylabel('y')
plt.legend(), plt.grid(True), plt.show()
```

- Інтерактивність: Здатність до взаємодії з введеним кодом та отримання миттєвого результату.  


```python
def plot_sine_wave(n):
    x = np.linspace(0, 2*np.pi, 100)
    y = np.sin(n * x)
    
    plt.figure(figsize=(8, 4))
    plt.plot(x, y, label=f'sin({n}x)')
    plt.xlabel('x'), plt.ylabel('y')
    plt.legend(), plt.grid(True), plt.show()

interact(plot_sine_wave, n=(1, 10, 1))
```

<p style="font-size: 20px;"> Вирахувати математичне сподівання та дисперсію для набору даних. Стандартизувати дані.

```python
# Вибірка чотирьох стовпців
selected_columns = ['Price', 'Mileage(miles)', 'Registration_Year']
sample_df = df[selected_columns]

# Розрахунок математичного сподівання та дисперсії для кожного стовпця
mean_values = sample_df.mean()
variance_values = sample_df.var()

# Стандартизація даних (Z-перетворення)
standardized_df = (sample_df - mean_values) / sample_df.std()

# Вивід результатів
print("Математичне сподівання:")
print(mean_values)

print("Дисперсія:")
print(variance_values)

print("Стандартизовані дані:")
print(standardized_df)
```

### Будуємо heatmap.

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
numeric_data = df[numeric_cols]

# Створення heatmap
plt.figure(figsize=(12, 8))
heatmap = sns.heatmap(numeric_data.corr(), annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

### Будуємо histplot діаграму.

```python
df['car_name'] = df['title'].str.split().str[0]

plt.figure(figsize=(14, 6))
sns.histplot(df['car_name'], bins=len(df['car_name'].unique()), kde=False, color='skyblue')
plt.xlabel('Brand')
plt.ylabel('Count (log scale)')
plt.xticks(rotation=90, ha='right')
plt.yscale('log')  # Встановлюємо логарифмічний масштаб для осі y
plt.show()
```
