    4.2 Вирахувати математичне сподівання та дисперсію для набору даних. Стандартизувати дані.

Математичне сподівання (μ):

$ \mu = \frac{1}{N} \sum_{i=1}^{N} x_i $


Дисперсія ($σ^2$):

$ \sigma^2 = \frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^2 $


````python
df['Engine'] = df['Engine'].str.extract('(\d+\.\d+)').astype(float)
````

````python
с_df = df[['Price', 'Mileage(miles)', 'Registration_Year', 'Engine']]

mean_values = с_df.mean(numeric_only=True)
variance_values = с_df.var(numeric_only=True)

standardized_df = (с_df - mean_values) / с_df.std()

standardized_df_1 = standardized_df.copy()
# Вивід результатів
print("Математичне сподівання:")
print(mean_values)

print("Дисперсія:")
print(variance_values)

print("Стандартизовані дані:")
print(standardized_df)
````

    6.1 Збудувати лінійний графік з 4-ма лініями, які взяті з ознак та додати annotate для середнього значення кожної лінії.

````python
selected_brands = ['Volkswagen', 'Ford', 'Audi', 'BMW']

selected_data = df[df['Brand'].isin(selected_brands)]

plt.figure(figsize=(16, 8))

sns.lineplot(x='Registration_Year', y='Price', hue='Brand', data=selected_data, palette="colorblind")

plt.legend()
plt.xlabel('Registration_Year')
plt.ylabel('Price')

for brand in selected_brands:
    mean_price = selected_data[selected_data['Brand'] == brand]['Price'].mean()
    max_mileage = selected_data[selected_data['Brand'] == brand]['Registration_Year'].max()

    line_color = sns.color_palette("colorblind", n_colors=len(selected_brands))[selected_brands.index(brand)]

    plt.annotate(f'Mean: {mean_price:.2f}', xy=(max_mileage, mean_price), color=line_color, fontsize=10,
                 ha='right', va='center')

plt.grid(True, linestyle='--', alpha=0.6, linewidth=0.5)
plt.show()
````

<p style="font-size: 18px;">7. Вирахувати коефіцієнт пірсона для кожної пари ознак.</p>


**Коефіцієнт кореляції Пірсона** (Pearson correlation coefficient) - це міра лінійної залежності між двома випадковими величинами. 

    7.1 Вирахувати коефіцієнт пірсона для кожної пари ознак.


````python
df_copy_3 = df_copy.copy()
correlation_matrix = df_copy_3.corr(method='pearson', numeric_only=True)
correlation_matrix 
````

````python
masked_matrix = np.ma.masked_equal(correlation_matrix.values, 1)

max_correlation = masked_matrix.max()
max_corr_indices = np.unravel_index(np.argmax(masked_matrix), masked_matrix.shape)

feature_x = correlation_matrix.columns[max_corr_indices[1]]
feature_y = correlation_matrix.index[max_corr_indices[0]]

print(f"Максимальна кореляція без урахування 1: {max_correlation}")
print(f"Ознаки з найбільшою кореляцією: {feature_x}, {feature_y}")
````

    7.2 Взяти третю ознаку за вісь X та збудувати relplot для двох ознак, що мають найбільшу кореляцію за допомогою бібліотеки seaborn.
