<p style="font-size: 22px;">4. Відображення 10 елементів для 4-ох ознак. Математичне сподівання та дисперсія. Стандартизування даних

<p style="font-size: 17px;">4.2 Вирахувати математичне сподівання та дисперсію для набору даних. Стандартизувати дані.</p>

Математичне сподівання (μ):

$ \mu = \frac{1}{N} \sum_{i=1}^{N} x_i $


Дисперсія ($σ^2$):

$ \sigma^2 = \frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^2 $


```python
# Вибірка чотирьох стовпців
sample_df = df[['Price', 'Mileage(miles)', 'Registration_Year']]

# Розрахунок математичного сподівання та дисперсії для кожного стовпця
mean_values = sample_df.mean()
variance_values = sample_df.var()

standardized_df = (sample_df - mean_values) / sample_df.std()

# Вивід результатів
print("Математичне сподівання:")
print(mean_values)

print("Дисперсія:")
print(variance_values)

print("Стандартизовані дані:")
print(standardized_df)
```
