```python
import pandas as pd
from datetime import datetime
import urllib
import os
print("Все ок)")
```

```python
date = ''
def time_now():
    now = datetime.now()
    global date 
    date = now.strftime('%d%m%Y%H%M%S')
```

```python
def download_data():
    indexes = ['', 22, 24, 23, 25, 3, 4, 8, 19, 20, 21, 9, 90, 10, 11, 12, 13, 14, 15, 16, 250, 17, 18, 6, 1, 2, 7, 5]

    time_now()
        
    if os.path.isdir('data'):
        for f in os.listdir('data'):
            os.remove('data/'+f)
    else:
        os.mkdir('data')
            
    for i in range(1,28):
        urls = 'https://www.star.nesdis.noaa.gov/smcd/emb/vci/VH/get_TS_admin.php?provinceID={}&country=UKR&yearlyTag=Weekly&type=Mean&TagCropland=crop&year1=1982&year2=2023'.format(i)
        text = urllib.request.urlopen(urls).read().decode('utf-8')
        
        out = open(f'data/NOAA_ID_{str(indexes[i])}_{date}.csv','wb')

        text = '\n'.join(text.split("\n")[1:]).replace("<tt><pre>", '').replace("<br>", '').replace("</pre></tt>", '').replace(",", ';').encode()
        out.write(text)
        out.close()
        print('0_0')
        
    print('Завантажилось!!!')
    
download_data()
```

```python
def dataframes(reg_index):
    df = pd.read_csv(f'data/NOAA_ID_{reg_index}_{date}.csv',
                     sep=';', index_col=False)
    df.columns = df.columns.str.replace(' ', '')
    df = df.drop(df.loc[df['VHI'] == -1].index)
    return df
```

```python
def min_max(region, year):
    df = dataframes(region)
    year_df = df[df['year'] == year]
    min_val = df[df['VHI'] == year_df['VHI'].min()]
    max_val = df[df['VHI'] == year_df['VHI'].max()]
    return min_val, max_val
min_max(13, 2000)
```

```python
def drought15(region, percent):
    df = dataframes(region)
    return df[ (df['VHI'] < 15) & (df['SMN'] > percent / 100) ]

drought15(2, 2)
```

```python
def drought35(region, percent):
    df = dataframes(region)
    return df[ (df['VHI'] > 15) & (df['VHI'] < 35) & (df['SMN'] > percent / 100) ]

drought35(2, 40)
```
