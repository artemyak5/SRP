```python
import pandas as pd
import timeit
import numpy as np

print('оК')
```

2. Таймер для порівняння швидкості


```python
def timer(function, number=1, repeat=1):
    execution_times = timeit.repeat(function, number=number, repeat=repeat)
    average_execution_time = sum(execution_times) / len(execution_times)
    print(f"Середній час виконання: {average_execution_time:.6f} секунд")
```

3. Зчитуємо дані


```python
def read_file():
    df = pd.read_csv('household_power_consumption.txt', index_col=None, header=0, sep=";", low_memory=False, na_values = '?')
    return df
read_file()
```

3. Обрати всі домогосподарства, у яких сила струму лежить в межах 19-20 А, для них виявити ті, у яких пральна машина та холодильних споживають більше, ніж бойлер та кондиціонер.</p>

```python
def pand_func_3():
    hz = pandas_df[(pandas_df['Global_intensity'] >= 19) & (pandas_df['Global_intensity'] <= 20)]
    return hz[hz['Sub_metering_2'] > hz['Sub_metering_3']]
pand_func_3()
```

4. Обрати випадковим чином 500000 домогосподарств (без повторів елементів вибірки), для них обчислити середні величини усіх 3-х груп споживання електричної енергії.</p>

```python
def pand_func_4():
    sampled_pandas_df = pandas_df.sample(500000)    
    sampled_pandas_df['Mean_of_3'] = sampled_pandas_df[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].mean(axis=1)    
    return sampled_pandas_df
pand_func_4()
```

```python
def nump_func_4():
    sampled_numpy_df = numpy_df[np.random.choice(numpy_df.shape[0], size=500000, replace=False), :]
    
    sub_metering_columns = sampled_numpy_df[:, [4, 5, 6]]  
    mean_values = np.mean(sub_metering_columns, axis=1) 
    sampled_numpy_df = np.column_stack((sampled_numpy_df, mean_values))
    
    return sampled_numpy_df

nump_func_4()
```

```python
timer(pand_func_4, repeat = 10) #Pandas
timer(nump_func_4, repeat = 10) #NumPy
```

<p style="font-size: 16px;">
5. Обрати ті домогосподарства, які після 18-00 споживають понад 6 кВт за хвилину в середньому, серед відібраних визначити ті, у яких основне споживання електроенергії у вказаний проміжок часу
припадає на пральну машину, сушарку, холодильник та освітлення (група 2 є найбільшою), а потім обрати кожен третій результат із першої половини та кожен четвертий результат із другої половини.</p>

```python
import pandas as pd

def pand_func_5():
    time_filtered_df = pandas_df[pandas_df['Time'] > pd.to_datetime("18:00:00", format='%H:%M:%S')]
    filtered_power_df = time_filtered_df[(time_filtered_df['Global_active_power'] > 6) & 
                                         (time_filtered_df["Sub_metering_2"] > time_filtered_df["Sub_metering_1"]) & 
                                         (time_filtered_df["Sub_metering_2"] > time_filtered_df["Sub_metering_3"])]
    
    first_third_subset = filtered_power_df.iloc[::3]
    second_fourth_subset = filtered_power_df.iloc[1::4]
        
    return pd.concat([first_third_subset, second_fourth_subset])
pand_func_5()
```

```python
numpy_df[:, 0] = numpy_df[:, 0].astype('U8')
```

```python
def nump_func_5():
    time_filtered_np = numpy_df[(numpy_df[:, 0]>'18:00:00')]
    filtered_power_np = time_filtered_np[(time_filtered_np[:, 2] > 6) & 
                                         (time_filtered_np[:, 7] > time_filtered_np[:, 6]) &
                                         (time_filtered_np[:, 7] > time_filtered_np[:, 8])]
    
    midpoint = filtered_power_np.shape[0] // 2
    first_third_subset = filtered_power_np[:midpoint:3]
    second_fourth_subset = filtered_power_np[midpoint::4]
    
    return np.concatenate((first_third_subset, second_fourth_subset), axis=0)

nump_func_5()
```

```python
timer(pand_func_5, repeat = 10) #Pandas
timer(nump_func_5, repeat = 10) #NumPy
```
